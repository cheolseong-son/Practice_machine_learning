## KNN(K-Nearest Neighbors)

#### 1) 특징
    - 이해하기 쉽고 직관적인 모델
    - 복잡한 알고리즘을 적용하기 전에 시도해 볼 수 있는 알고리즘
    - 훈련셋이 너무 크면 예측이 느려짐
    
#### 2) 파라미터
    - 이웃의 갯수(K)
    - 데이터 사이의 거리를 재는 방법 : L2 norm
    - L1 : 단순한 거리(맨하탄 거리), L2 : 최단 거리(피타고라스, 유클리드)

#### L1 norm, L2 norm

- L1 norm 
        - 단순한 거리(맨하탄 거리)의 절대값의 합

- L2 norm
        - 최단 거리(피타고라스, 유클리드)
        - 거리의 절대값의 합에 루트 적용

```python
# 유방암 데이터로 실습

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
cancer
# 훈련용과 테스트용으로 분류
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=6) # random_state=0 : 난수 고정
print(X_train.shape, X_test.shape)
```

```python
clf = KNeighborsClassifier(n_neighbors = 6).fit(X_train, y_train)
clf.score(X_train, y_train) # 100% 일치
```

```python
clf.score(X_test, y_test) # 91% 일치
# n_neighbors의 이웃의 숫자가 절적한 숫자로 올라갈수록 test데이터 훈련에서 정확도가 올라간다.
```

```python
# 최적의 이웃의 숫자를 구하기 위해 반복문 사용

train_acc = []
test_acc = []
for n in range(1,20):
    clf = KNeighborsClassifier(n_neighbors = n).fit(X_train, y_train)
    train_acc.append(clf.score(X_train, y_train)) 
    test_acc.append(clf.score(X_test, y_test)) 
```

```python
# 그래프로 그리기

plt.plot(range(1,20), train_acc, label = "훈련 정확도")
plt.plot(range(1,20), test_acc, label = "테스트 정확도")
# 훈련데이터와 테스트 데이터의 접점이 가장 최적의 갯수
plt.xlabel("정확도")
plt.ylabel("이웃의 개수")
plt.legend()
```

#### 영화 분류 프로그램 만들기


```python
import operator

def createDataSet():
    group = np.array([[1.0, 2.0],[1.0,4.0],[4.0,1.0],[4.0,2.0]]) # 학습 데이터 2차원 4행 2열
    label = ["Action", "Action", "Romance", "Romance"]
    return group, label

# 거리계산 함수 만들기
def clacDistance(X_new, data):
    # 데이터 개수
    dataSize = len(data) # 또는 data.shape[0]
    
    # 입력 데이터와 기존 데이터의 차이
    diffMat = np.tile(X_new, (dataSize,1)) - data
    
    # L2 norm 거리계산
    sqDiffMat = diffMat**2
    sqDistance = sqDiffMat.sum(axis = 1)
    distance = sqDistance ** 0.5
    print(distance)
    
    # 정렬된 위치
    return distance.argsort()


def findMajorityClass(target, k, sortedDistance): # 가장 가가운게 액션인지 로맨스인지 알려주는 함
    classCount = {}
    
    for i in range(k):
        # 로맨스, 액션 구분
        # target에는 액션, 로맨스 값이 지정되어 있다.
        voteLabel = target[sortedDistance[i]]
        # print(voteLable)
        classCount[voteLabel] = classCount.get(voteLabel, 0) + 1 # 0은 처음 값이 없으면 0부터 시작하고  +1을 한다
    # print(classCount)
        # dict를 list or tuple로 만들어 주기 정렬을 하기 위해
    return sorted(classCount.items(), key=operator.itemgetter(1), reverse = True) 
        # key=operator.itemgetter(1) :: 첫번째 값을 정렬기준으로, 0 이면 알파벳순 정렬
        # reverse = True : 내림차순 정렬
        
def classify(X_new, data, target, k): # 인자 4개를 받는 함수, k : 이웃 개수
    sortedDistance = clacDistance(X_new, data) # 거리계산을 변수에 저장
    sortedDistanceCount = findMajorityClass(target, k, sortedDistance)
    
    return sortedDistanceCount
```

```python
X, y = createDataSet()
result = classify([2.0, 3.0], X, y, 3)
print("결과는 {}영화다.".format(result[0][0]))

# 가장 가까운 곳의 인덱스 값은 첫번째 먼 곳은 3번째 
```
